```python
%pylab inline
import cv2
import os
import PyOpenPose as OP
OPENPOSE_ROOT = os.environ["OPENPOSE_ROOT"]
```

```python
img = cv2.imread("galloping knights.jpg")
```

```python
op = OP.OpenPose((656, 368), (368, 368), (1280, 720), "COCO", OPENPOSE_ROOT + os.sep + "models" + os.sep, 0, False)
```

```python
op.detectPose(img)
```
